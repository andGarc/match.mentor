```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# dataframe
mentors = ['Mentor1', 'Mentor2', 'Mentor3']
mentees = ['Mentee1', 'Mentee2', 'Mentee3']
confidence_values = [
    [0.8, 0.6, 0.9],
    [0.7, 0.5, 0.8],
    [0.9, 0.8, 0.95]
]
df = pd.DataFrame(confidence_values, index=mentors, columns=mentees)
df
```

```python
# Create a heatmap using seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(df, annot=True, cmap='YlGn', fmt=".2f", cbar_kws={'label': 'Confidence (%)'})

# Set axis labels and plot title
plt.xlabel('Mentees')
plt.ylabel('Mentors')
plt.title('Mentor-Mentee Confidence Heatmap')

# Show the plot
plt.show()
```
